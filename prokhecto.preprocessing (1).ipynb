```python
tsdf=[]

for tspath in glob("timeseries/*.json"):
    with open(tspath, 'r', encoding='utf8') as f:
        try:
            tsdf+=json.load(f)
        except:
            continue

tsdf=pd.DataFrame(tsdf)
#to preprocess timeseries->transactions
tsdf = tsdf.groupby(['symbol']).mean().reset_index()#kherjena symbol men index
#calcul dyal mean rate per hour
#9.904218e-04--<9.9 X 10^-4

tsdf.info()
tsdf
```

```python
metadata=[]
urlsList=[]
for tspath in glob("metadata/*.json"):
    with open(tspath, 'r', encoding='utf8') as f:
        try:
            x=json.load(f)
            U=[a.get("urls", {}) for a in x]
            urlsList+=U
            metadata+=x
        except Exception as e:
            raise(e)
            continue

metadata=pd.DataFrame(metadata).drop("urls",axis=1)

metadata.info()
#to preprocess urls & other attributes
```

```python
#REGEX
import re

url = "https://alphalabz.net/"
name = "Alpha Labz"

def find_name_in_url(urls, name):
    name = re.sub('\W', '', name)
    return all([len(re.findall(name, url, flags= re.IGNORECASE)) != 0 for url in urls])
#sum=bool>>int
```

```python
urlsDf=pd.DataFrame(urlsList)
# urlsDf = urlsDf.apply(len)
for c in urlsDf.columns:
    #urlsDf[c]=urlsDf[c].apply(len)
#replace a column with its lenght

    metadata[c]=urlsDf[c].apply(len)
#to add urls to metdata
urlsDf["id"]=metadata["id"]
urlsDf["name"]=metadata["name"]

#Verifies if url contains name of coin
# urlsDf.apply(lambda x: find_name_in_url(x["website"],x["name"]), axis=1)
# urlsDf.tail()[["website","name"]]

urlsDf

# #regex
# #website = "The rain in Spain"
# x = re.findall("ai", txt)



# print("The first white-space character is located in position:", x.start())
```

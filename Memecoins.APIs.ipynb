```python
def solve_dict(D):
    non_dict_val = {k:v for k, v in D.items() if type(v) != dict}
    dict_val = {k:v for k, v in D.items() if type(v) == dict}
    dict_val = {k+'_'+ke: ve for k, v in dict_val.items() for ke, ve in v.items()}
    #Dict in dict

    non_dict_val.update(dict_val)
    return non_dict_val
```

```python
[solve_dict(d) for d in content]
```

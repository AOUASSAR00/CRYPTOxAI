```python
tsdf=[]

for tspath in glob("timeseries/*.json"):
    with open(tspath, 'r', encoding='utf8') as f:
        try:
            tsdf+=json.load(f)
        except:
            continue

tsdf=pd.DataFrame(tsdf)
#to preprocess timeseries->transactions
tsdf = tsdf.groupby(['symbol']).mean().reset_index()#kherjena symbol men index
#calcul dyal mean rate per hour
#9.904218e-04--<9.9 X 10^-4

tsdf.info()
tsdf
```

```python
metadata=[]

for tspath in glob("metadata/*.json"):
    with open(tspath, 'r', encoding='utf8') as f:
        try:
            metadata+=json.load(f)
        except:
            continue

metadata=pd.DataFrame(metadata)
metadata['website'] = metadata['urls'].apply(lambda x: x.get("website"))
metadata.info()
#to preprocess urls & other attributes
```
